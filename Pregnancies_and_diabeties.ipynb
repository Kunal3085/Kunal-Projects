```python
# Univariate Analysis#
# Continues variables (distribution plot)
continues_variable = ['Glucose', 'BloodPressure','Insulin','BMI','Age']
for var in continues_variable:
  plt.figure(figsize=(8,5))
  sns.histplot(df[var], kde=True, bins=30, color='skyblue')
  plt.title(f'Distribution of {var}')
  plt.xlabel(var)
  plt.ylabel('Frequency')
  plt.show()
```

```python
# Bivariate anlysis relations between Pregnancies	and  Glucose
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.boxplot(x='Pregnancies', y='Glucose', data=df)
plt.xlabel('Pregnancies')
plt.ylabel('Glucose')
plt.title('Relations between Pregnancies and Glucose')
plt.show()
```

```python
#Bivariate analysis between Pregnancies and  Age
sns.boxplot(x='Pregnancies', y='Age', data=df)
plt.xlabel('Pregnancies')
plt.ylabel('Age')
plt.title('Relations between Pregnancies and Age')
plt.show()
```

```python
#Bivariate analysis between Pregnancies and  BMI
sns.boxplot(x='Pregnancies', y='BMI', data=df)
plt.xlabel('Pregnancies')
plt.ylabel('BMI')
plt.title('Relations between Pregnancies and BMI')
plt.show()
```

```python
#Identify Outliers using Boxplots
#Boxplot for Pregnancies
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Pregnancies'], color='green')
plt.title('Boxplot for Pregnancies')
plt.xlabel('Pregnancies')
plt.show()
# There are three Outliners for Pregnancies#
```

```python
#Identify Outliers using Boxplots
#Boxplot for Pregnancies Glucose
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Glucose'], color='red')
plt.title('Boxplot for pregnacies Glucose level')
plt.xlabel('Glucose')
plt.show()
# There are only one outliners for Pregnancies Glucose level#
```

```python
#Identify Outliers using Boxplots#
#Boxplot for Pregnancies Bloodpressure#
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(8, 5))
sns.boxplot(x=df['BloodPressure'], color='blue')
plt.title('Boxplot for pregnacies BloodPressure')
plt.xlabel('BloodPressure')
plt.show()
# There are lot of outliners in the Pregnancies Bloodpressure#
```
